```python
url = 'https://raw.githubusercontent.com/wtm695450085/Perfect-Plots__GIP/master/diabetes.csv'
df = pd.read_csv(url)
df.head(2)
```

```python
df.groupby(['SkinThickness', 'Age'])['Age'].count().to_frame().unstack().fillna("")
```

```python
import matplotlib.pyplot as plt  
import seaborn as sns

df2 = df.groupby(['SkinThickness', 'Age'])['Age'].count().to_frame().unstack()
plt.figure(figsize=(13,3))
sns.heatmap(df2, cmap="Reds", annot=True, cbar=False)
```

# lambda

```python
url = 'https://raw.githubusercontent.com/wtm695450085/Perfect-Plots__GIP/master/uforeports.csv'
df = pd.read_csv(url)
df.head(2)
```

```python
df[df['Shape Reported']=='FIREBALL'].groupby('City').filter(lambda g: len(g) > 1).groupby('City').size().sort_values(ascending=False)
```
