```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/wtm695450085/Perfect-Plots__GIP/master/autos.csv")
df.head(2)
```

# ile jakich rodzaj√≥w nadwozia

```python
pd.crosstab(df.make, df.body_style)
```

```python
KD= pd.crosstab(df.make, df.body_style.count())
KD.loc['Razem']=KD.sum(axis=0)
KD
```

```python
pd.crosstab(df.make, df.body_style, margins=True, margins_name= 'Razem')
```
