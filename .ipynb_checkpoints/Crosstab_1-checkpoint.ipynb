```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/wtm695450085/Perfect-Plots__GIP/master/bank.csv")
df.head(2)
```

```python
pd.crosstab(index=df.marital,
            columns=df.y)
```

# procentowy udzia≈Ç

```python
table = pd.crosstab(df.marital, df.y)
table.div(table.sum(1).astype(float), axis=0)
```

```python
pd.crosstab(df.marital, df.y)

table.div(table.sum(1).astype(float), axis=0).plot(kind='bar')
plt.title('Stacked Bar Chart of Marital Status vs Purchase')
plt.xlabel('Marital Status')
plt.ylabel('Proportion of Customers')
plt.savefig('mariral_vs_pur_stack')
```

```python
pd.crosstab(df.marital, df.y)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Marital Status vs Purchase')
plt.xlabel('Marital Status')
plt.ylabel('Proportion of Customers')
plt.savefig('mariral_vs_pur_stack')
```

```python
table = pd.crosstab(df.education, df.y)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Chart of Education vs Purchase')
plt.xlabel('Education')
plt.ylabel('Proportion of Customers')
plt.savefig('edu_vs_pur_stack')
```

```python
table = pd.crosstab(df.day_of_week, df.y)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Day fo week vs Purchase')
plt.xlabel('Day fo week')
plt.ylabel('Proportion of Customers')
plt.savefig('Day fo week')
```

```python
table = pd.crosstab(df.day_of_week, df.y).plot(kind='bar', stacked=True)
plt.title('Day fo week vs Purchase')
plt.xlabel('Day fo week')
plt.ylabel('Proportion of Customers')
plt.savefig('Day fo week')
```

```python
pd.crosstab(df.month,df.y).plot(kind='bar')
plt.title('Purchase Frequency for Month')
plt.xlabel('Month')
plt.ylabel('Frequency of Purchase')
plt.savefig('pur_fre_month_bar')
```

```python
table = pd.crosstab(df.age, df.y)
table.div(table.sum(1).astype(float),
          axis=0).plot(kind='bar', stacked=True, figsize=(26, 4))

plt.title('Stacked Bar Chart of age vs Purchase')
plt.xlabel('Age')
plt.ylabel('Proportion of Customers')
plt.savefig('Age_vs_pur_stack')
```

```python
table = pd.crosstab(df.age, df.y).plot(kind='bar',
                                      stacked=False, figsize=(26, 4))

plt.title('Stacked Bar Chart of age vs Purchase')
plt.xlabel('Age')
plt.ylabel('Proportion of Customers')
plt.savefig('Age_vs_pur_stack')
```

```python
pd.crosstab(df.poutcome, df.y).plot(kind='bar')
plt.title('Purchase Frequency for Poutcome')
plt.xlabel('Poutcome')
plt.ylabel('Frequency of Purchase')
plt.savefig('pur_fre_pout_bar')
```
